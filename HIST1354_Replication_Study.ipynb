##Importing Files

 Below is the code I used to import the .csv file that the original study used. To do this, I first uploaded the document into Coop and then wrote the file’s path into the code. Next, I used pandas to extract the data in the .csv and, to esnure everything had run smoothly, I had the program print the data.

```python
file_path = ('/content/Synagogues_of_Massachusetts,_By_Decade_data.csv')
df = pd.read_csv(file_path)
df.head(500)
```

##Separating and Counting the Data

Below is the code I used to have the computer separate the data into decades so that I could later map it in a timeline. I did this by telling the code to group the data frame by the “Decade Founded” column in the data table, count the number of rows in each decade, and then report that information back to me in a table with the decade itself and its corresponding count. I then told the code to print me said chart to ensure everything was executed properly.

```python
timeline_data = df.groupby('Decade Founded').size().reset_index(name='Counts')

timeline_data.head(17)
```

##Creating the Visualisation

Now that the data has been organized into decades and then counted, I am able to create a graph that visualizes this data. To do this, I primarily relied on seaborn and matplotlib. I first set the paramotors of the graph (ie: a 10x10 graph with a white background), and then told it which data I wanted it to chart and how I wanted it to be charted. This is where I ran into some issues.  

My first attempt resulted in the graph displaying the data correctly, but in numerical order rather than chronological order. To fix this, I added additional syntax to ensure the code knew that I wanted the graph to be organized chronologically. Unfortunately, the graph then displayed the data with the most recent decade on the left side of the graph whereas the study had it where the least recent data was on the left. To better replicate the study, I then added the “sorted().unique()” methods to the code which solved the issue.  

Lastly, using matplotlib, I added the graph and axis titles and then told the graph to print out.

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,5))

sns.countplot(data=df, x='Decade Founded', order=sorted(df['Decade Founded'].unique()))

plt.title('Number of Newly Founded Synagogues in Boston, by Decade', fontsize=16)
plt.xlabel('Decade')
plt.ylabel('Number of New Synaogogues Founded')
plt.show()
```
